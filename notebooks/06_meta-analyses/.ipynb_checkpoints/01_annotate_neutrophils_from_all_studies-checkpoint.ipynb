# Concat untreated MC38 datasets

```python
mc38_outer = sc.concat((sdata_tumor_nt, gdata_nt), join='outer')
mc38_inner = sc.concat((sdata_tumor_nt, gdata_nt), join='inner')

sc.pp.normalize_total(mc38_outer, 1e4)
sc.pp.normalize_total(mc38_inner, 1e4)
```

```python
datasets = {'Gungabeesoon KP19 untreated': ndata,
             'Gungabeesoon KP19 aCD40': tdata,
             'Garris MC38 aPD1': gdata_apd1,
             'Siwicki MC38 aCD40': sdata_tumor_acd40,
             'MC38 untreated':mc38_outer}

for i in datasets.keys():
    datasets[i] =  datasets[i][datasets[i].obs['N_label'].isin(datasets[i].obs['N_label'].value_counts()[datasets[i].obs['N_label'].value_counts() >= 10].index)].copy()
```

```python
centroids = {dataset_name: ng.expression_centroids(d, 'N_label') for dataset_name, d in datasets.items()}
```

```python
clf_matrices = {}

for i in tqdm(datasets.keys()):
    clf_matrices[i] = {}
    for j in datasets.keys():
        clf_matrices[i][j] = ng.parallel_cell_classifier(adata = datasets[i], bdata = datasets[j], state_column_b = 'N_label', 
                                                                comment=f'{ng.fix_filename(i)}_neutrophils_by_{ng.fix_filename(j)}', 
                                                                save_dir = '/n/groups/klein/nico/neutrophils/backups/', progress_bar = False,
                                                                step = np.ceil(len(datasets[i])/4).astype(int), return_most_likely=False, nproc=4).T
        
        clf_matrices[i][j].index = datasets[i].obs_names
```

```python
label_correspondences = {}
for i in tqdm(datasets.keys()):
    label_correspondences[i] = {}
    for j in datasets.keys():
        
        corresp = ng.label_correspondence(datasets[i], 'N_label', clf_matrices[i][j].reset_index(drop=True).T)
        corresp.index = corresp.index.astype(str) + f'_{ng.fix_filename(i)}'
        corresp.columns = corresp.columns.astype(str) + f'_{ng.fix_filename(j)}'
        
        label_correspondences[i][j] = corresp
```

```python
mutual_correspondences = {}
for i in tqdm(datasets.keys()):
    mutual_correspondences[i] = {}
    for j in datasets.keys():
        mutual_correspondences[i][j] = label_correspondences[i][j] * label_correspondences[j][i].T
```

```python
rss_mtx = pd.concat([pd.concat(mutual_correspondences[i].values(), axis=1) for i in datasets.keys()],axis=0)

g = sns.clustermap(np.sqrt(rss_mtx), figsize=(16,16), 
              cmap = 'coolwarm', linewidths = .5, vmin = 0, vmax = 1, tree_kws={'linewidth':2}, cbar_pos=(1.02,0.3, 0.02,0.4), cbar_kws={'label':''})
g.ax_heatmap.set_ylabel("")
plt.gcf().set_dpi(200)
```
