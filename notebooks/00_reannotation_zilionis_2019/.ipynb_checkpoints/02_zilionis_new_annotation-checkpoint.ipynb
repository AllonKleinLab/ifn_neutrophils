# Load re-embedded data

```python
zdata = sc.read('/n/groups/klein/nico/neutrophils/backups/zilionis_mouse_reembedded_15939x27045_backup_220322_13h54.h5ad')
```

# Leiden

Find the smallest cluster that includes cell *t_1_3_bc0057* (Ngp++ Camp++)

```python
for res in [0.5, 1, 2, 4, 6, 8, 10, 12]:
    sc.tl.leiden(zdata, res, key_added=f'leiden{res}')
    ng.umap_plot(zdata, color=f'leiden{res}', groups=zdata['t_1_3_bc0057'].obs[f'leiden{res}'][0], basis='X_SPRING')
    print(f'{res}:',(zdata.obs[f'leiden{res}'] == zdata['t_1_3_bc0057'].obs[f'leiden{res}'][0]).sum())
```

# Split N1s into N1a and N1b depending on leiden8

Reannotate cells from the cluster *t_1_3_bc0057* is part of as N1a

```python
zdata.obs['state_label'] = zdata.obs['Minor subset'].astype('str')

zdata.obs.loc[((zdata.obs['leiden8'] == '101') & (zdata.obs['Minor subset'] == 'N1')).astype('category'), 'state_label'] = 'N1a'
zdata.obs.loc[((zdata.obs['leiden8'] != '101') & (zdata.obs['Minor subset'] == 'N1')).astype('category'), 'state_label'] = 'N1b'

zdata.obs['state_label'] = zdata.obs['state_label'].astype('category')
```

# Save

```python
start = time()
fname = '/n/groups/klein/nico/neutrophils/backups/zilionis_mouse_reannotated_{}x{}_backup_{}.h5ad'.format(*zdata.shape,ng.now())
zdata.write(fname)
print(fname)
ng.print_etime(start)
```
