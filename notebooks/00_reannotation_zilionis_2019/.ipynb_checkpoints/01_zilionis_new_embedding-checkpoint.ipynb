# Load data from reference

```python
zdata = sc.read_mtx("/n/groups/klein/nico/neutrophils/data/zilionis19/GSE127465_mouse_counts_normalized_15939x28205.mtx.gz")
zdata.var_names = pd.read_csv("/n/groups/klein/nico/neutrophils/data/zilionis19/GSE127465_gene_names_mouse_28205.tsv.gz", sep='\t', index_col=0, header=None).index.rename(None)
zdata.obs = pd.read_csv("/n/groups/klein/nico/neutrophils/data/zilionis19/GSE127465_mouse_cell_metadata_15939x12.tsv.gz", sep='\t')
zdata.obsm['X_SPRING'] = zdata.obs.loc[:,['x', 'y']].values

zdata._sanitize()
zdata.obs_names = (zdata.obs['Library'].astype(str) + '_' + zdata.obs['Barcode'].astype(str))

from scipy.sparse import csr_matrix
zdata.X =  csr_matrix(np.round(np.asarray(zdata.X /zdata.X.sum(1)) *  zdata.obs['Total counts'][:, None]))
zdata.raw = zdata # save UMI in raw

# Match rescaling
# zdata = ng.sctransform(zdata, hvg=False, rlib_loc="/n/groups/klein/nico/neutrophils/analysis/conda-env/lib/R/library", min_cells=3, copy=True, n_genes=8000, verbose=False)
# zdata.layers['cp10k'] = sc.pp.normalize_total(zdata, target_sum=1e4, inplace=False)['X']

# start = time()
# fname = '/n/groups/klein/nico/neutrophils/backups/zilionis19_mouse_cp10k_sct_{}x{}_backup_{}.h5ad'.format(*zdata.shape,ng.now())
# zdata.write(fname)
# print(fname)
# ng.print_etime(start)
```

# Expression layers

```python
start = time()

uns_backup = zdata.uns # remove uns to convert to Seurat object
del zdata.uns

# apply sct v1
zdata = ng.sctransform(zdata, hvg=False, rlib_loc="/n/groups/klein/nico/neutrophils/analysis/conda-env/lib/R/library", min_cells=3, 
                       copy=True, n_genes=8000, verbose=False, correct_umi=True, v2_flavor=False)

# norm cp10k
zdata.layers['cp10k'] = sc.pp.normalize_total(zdata, target_sum=1e4, inplace=False)['X']

# log(1 + cp10k)
zdata.layers['log1p'] = zdata.layers['cp10k'].copy()
sc.pp.log1p(zdata, layer='log1p')

# z_score log(1 + cp10k)
zdata.layers['scaled'] = zdata.layers['log1p'].copy()
sc.pp.scale(zdata, layer='scaled')

zdata.uns.update(uns_backup)

ng.print_etime(start)
```

# Dimensionality reduction, embedding

```python
pca_genes = zdata.var['highly_variable']
n_pcs = 50
zdata.uns['pca'] = {}
zdata.obsm['X_pca'], pca_loadings, zdata.uns['pca']['variance_ratio'], zdata.uns['pca']['variance'] = sc.tl.pca(zdata.layers['sct'][:,pca_genes],
                                                                                                                svd_solver = 'arpack', n_comps = n_pcs, return_info=True) 

ng.print_etime(start)

# expand PC loading matrix to include non-HVG rows filled with zeros
pca_loadings_all = (zdata.var.join(pd.DataFrame(pca_loadings.T, index=zdata.var_names[pca_genes]))).iloc[:,-n_pcs:].fillna(0).values

zdata.varm['PCs'] = pca_loadings_all

sc.pp.neighbors(zdata, n_neighbors=15, use_rep='X_pca') #default
ng.print_etime(start)
sc.tl.umap(zdata)
ng.print_etime(start)
```
