# Load data and metadata

```python
n_states = ['N1a', 'N1b', 'N2', 'N3', 'N4', 'N5', 'N6']
```

```python
n_states_dict = {'N1a' : 'Sell+Ngp+ neutrophils',
                 'N1b' : 'Sell+Lst1+ neutrophils',
                 'N2' : 'Sell+Cxcl10+ neutrophils',
                 'N3' : 'Cxcl3+ neutrophils',
                 'N4' : 'Siglecf+Xbp1+ neutrophils',
                 'N5' : 'Siglecf+Ccl3+ neutrophils',
                 'N6' : 'Siglecf+Ngp+ neutrophils'}

neu_palette_dict = {'Sell+Ngp+ neutrophils': '#6EA6CD',
                    'Sell+Lst1+ neutrophils': '#4A7BB7',
                    'Sell+Cxcl10+ neutrophils': '#364B9A',
                    'Cxcl3+ neutrophils': '#EAECCC',
                    'Siglecf+Xbp1+ neutrophils': '#A50026',
                    'Siglecf+Ccl3+ neutrophils': '#DD3D2D',
                    'Siglecf+Ngp+ neutrophils': '#F67E4B'}
```

```python
tdata = sc.read('/n/groups/klein/nico/neutrophils/backups/totalseq_exp3_neutrophils_acd40_annotated_embedding_25210x17824_backup_220422_15h46.h5ad')
```

```python
neutrotime_markers = 'Serpinb1a Wfdc21 Chil3 Dstn Camp Ly6c2 Arhgdib Anxa1 Adpgk Ngp Lcn2 Prdx5 Ltf Pglyrp1 Mgst1 Cd177 Cybb Pfn1 Ly6g Ifitm6 Mmp8 Lgals3 Retnlg Lyz2'.split()
neutrotime_markers = [g for g in neutrotime_markers if g in tdata.var_names]
```

# Heatmap - CP10K

```python
hmdata = ng.groupby_aggregate(tdata, layer='cp10k', groupby=['library_name','smoothed_Zilionis'], var_names= neutrotime_markers, f=np.mean, return_df=True).groupby('smoothed_Zilionis').mean()
hmdata = np.log2((hmdata + 1e-4) / (hmdata.median(0) + 1e-4))
```

```python
vmin = -4
vmax = 4

with sns.plotting_context({'figure.dpi':300}):
    sns.heatmap(hmdata.loc[:,(hmdata.T['N1a'] + hmdata.T['N6']).sort_values(ascending=False).index], cmap='coolwarm', square=True, 
                cbar_kws={'shrink':0.35, 'ticks':range(vmin, vmax+1,2)}, vmax=vmax, vmin=vmin, center=0)
    
    plt.gca().set_ylabel('')
    plt.gca().set_yticks([])
    plt.gca().tick_params(bottom=False)
```

# Use Z-scores for neutrotime score; flip sign

```python
exp_layer = 'scaled'

tdata.X = tdata.layers[exp_layer].copy()
sc.tl.score_genes(tdata, gene_list= neutrotime_markers, score_name='neutrotime_score', use_raw=False)
tdata.X = tdata.layers['umi'].copy()
```

```python
fig, ax = plt.subplots(1, figsize=(8,4), dpi=200)
sns.boxplot(data=tdata.obs, x=-tdata.obs['neutrotime_score'], y='smoothed_Zilionis', ax=ax, palette={k:neu_palette_dict[v] for k,v in n_states_dict.items()}, fliersize=0)
ax.set_xlim((-1.75,1.5))
```
