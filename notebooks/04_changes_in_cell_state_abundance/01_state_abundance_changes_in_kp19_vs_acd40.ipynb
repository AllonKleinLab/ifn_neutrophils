# Load neutrophils from experiments 2 and 3

```python
ndata = sc.read('/n/groups/klein/nico/neutrophils/backups/totalseq_exp2_neutrophils_untreated_annotated_embedding_5258x13126_backup_220422_15h46.h5ad')
tdata = sc.read('/n/groups/klein/nico/neutrophils/backups/totalseq_exp3_neutrophils_acd40_annotated_embedding_25210x17824_backup_220422_15h46.h5ad')
```

```python
hdata = ndata[ndata.obs.sample_condition.str.startswith('WT')].copy()
udata = ndata[ndata.obs.sample_condition.str.startswith('KP19')].copy()
```

* 12375: treated neutrophils per mg
* 5362: untreated neutrophils per mg (FACS)

## Compute untreated, treated, all abundances

```python
healthy_abundances = (hdata.obs.groupby(['library_name','smoothed_Zilionis']).size() / hdata.obs.groupby(['library_name']).size())
untreated_abundances = (udata.obs.groupby(['library_name','smoothed_Zilionis']).size() / udata.obs.groupby(['library_name']).size())
treated_abundances = (tdata.obs.groupby(['library_name','smoothed_Zilionis']).size() / tdata.obs.groupby(['library_name']).size())

all_abundances = pd.concat((healthy_abundances.reset_index(name='p').assign(sample_condition='Healthy'),
                            untreated_abundances.reset_index(name='p').assign(sample_condition='Untreated'), 
                            treated_abundances.reset_index(name='p').assign(sample_condition='aCD40')))#.melt(id_vars=['sample_condition', 'library_name', 'smoothed_Zilionis'], value_vars='p', )
```

# Adjust by change in absolute cell numbers

## Compute averages and SEM for each condition

```python
treated_abundances_abs = 12375 * treated_abundances
untreated_abundances_abs = 5362 * untreated_abundances


all_abundances_abs = pd.concat((untreated_abundances_abs.reset_index(name='p').assign(sample_condition='Untreated'), 
                               treated_abundances_abs.reset_index(name='p').assign(sample_condition='aCD40')))

abundance_stats_abs = all_abundances_abs.groupby(['sample_condition', 'smoothed_Zilionis']).agg(['mean', 'sem']).reset_index()
abundance_stats_abs = abundance_stats_abs.pivot_table(index='smoothed_Zilionis', columns='sample_condition')
abundance_stats_abs.columns = ['_'.join(col) for col in abundance_stats_abs.columns]
abundance_stats_abs = abundance_stats_abs.reset_index()
```

# Plot

```python
scatter_stats_abs = abundance_stats_abs[abundance_stats_abs['smoothed_Zilionis'].isin(n_states)]

fig, ax = plt.subplots(1, figsize=(4,4), dpi=200)

ax.errorbar(x=scatter_stats_abs['p_mean_Untreated'], y=scatter_stats_abs['p_mean_aCD40'], xerr= 1.96* scatter_stats_abs['p_sem_Untreated'], yerr=1.96 * scatter_stats_abs['p_sem_aCD40'], fmt='none', ecolor='k', elinewidth=0.4)

sns.scatterplot(data=scatter_stats_abs, x='p_mean_Untreated', y='p_mean_aCD40', hue='smoothed_Zilionis',
                    legend=False, ax=ax, zorder=2, s=80, palette= n_palette.values())


ax.set_xscale('log')
ax.set_yscale('log')


ax.set_xlabel(r'Tumor burdened ($\frac{cells}{mg})$')
ax.set_ylabel(r'aCD40-treated ($\frac{cells}{mg}$)')

ax.plot([0,1e4], [0,1e4], c='k', ls='--', alpha=0.3)
ax.plot([1e-4,1e4], [10e-4,10e4], c='r', ls='dotted', alpha=0.3, label='10-fold enrichment')
ax.plot([1e-4,1e4], [.1e-4,.1e4], c='b', ls='dotted', alpha=0.3, label='10-fold depletion')


ax.set_xlim([1e0,1e4])
ax.set_ylim([1e0,1e4])

plt.legend(fontsize=8, bbox_to_anchor=(1,1.02))
plt.show()
```
