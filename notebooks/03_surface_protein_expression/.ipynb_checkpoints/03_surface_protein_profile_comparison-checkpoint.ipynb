# Only untreated

```python
clr_averages_untreated = ng.groupby_aggregate(udata, ['smoothed_Zilionis', 'library_name'], layer='clr', f=np.mean, 
                                    return_df=True, var_names=ndata.var_names).dropna().groupby(['smoothed_Zilionis']).mean()

clr_medians_across_groups_untreated = clr_averages_untreated.median(0)
clr_second_max_across_groups_untreated = pd.DataFrame([clr_averages_untreated[col].sort_values(ascending=False).reset_index(drop=True) for col in clr_averages_untreated]).T.iloc[1]
clr_averages_min_untreated = clr_averages_untreated.min(0)

clr_averages_minus_min_untreated = clr_averages_untreated - clr_averages_min_untreated
clr_averages_minus_median_untreated = clr_averages_untreated - clr_medians_across_groups_untreated
```

```python
with plt.rc_context({'font.size':6}):
    fig = plt.figure(dpi=200, figsize=(8,2))
    sns.heatmap(clr_averages_minus_min_untreated, cmap='viridis', vmax=4)
    
    plt.gca().set_title('Untreated KP')
    plt.gca().tick_params(axis='x', rotation=45)
    plt.setp(plt.gca().get_xticklabels(), rotation=45, ha="right", rotation_mode="anchor")
    
    fig = plt.figure(dpi=200, figsize=(8,2))
    sns.heatmap(clr_averages_minus_median_untreated, cmap='coolwarm', vmin=-2, vmax=2)
    
    plt.gca().set_title('Untreated KP')
    plt.gca().tick_params(axis='x', rotation=45)
    plt.setp(plt.gca().get_xticklabels(), rotation=45, ha="right", rotation_mode="anchor")
    
    
```
